## Visualization & Explanatory Analysis

### Pertanyaan 1:

1. Berapa rata-rata total penyewa pada tiap musim(*season*)?. Berapa total penyewa casual dan total penyewa registered pada tiap-tiap season?. Total penyewa di season apa yang paling banyak dan season yang mana yang paling sedikit?


---
---


- kode dibawah akan menjawab rata-rata total penyewa pada tiap season

```python
season_calc = day_hari.groupby("season").agg({"cnt":"mean"})
season_calc.plot(kind="bar", figsize=(10, 6))
plt.title("rata-rata penyewa casual dan registered tiap season")
plt.xlabel("season")
plt.ylabel("jumlah penyewa")
plt.xticks(rotation=0)
plt.show()
```



---



---


- kode dibawah akan menjawab total penyewa casual dan total penyewa registered pada tiap-tiap season

```python
season_calc = day_hari.groupby("season").agg({"casual":"sum", "registered":"sum"})
season_calc.plot(kind="bar", figsize=(10, 6))
plt.title("penyewa casual dan penyewa registered tiap season")
plt.xlabel("season")
plt.ylabel("jumlah penyewa")
plt.xticks(rotation=0)
plt.show()
```



---



---

> *le6 pada tabel dibawah adalah notasi ilmiah pada ilmu komputer untuk angka besar yaitu = 1.000.000 (satu juta).*


- kode dibawah akan menampilkan penyewa di tiap season untuk mengetahui penyewa paling banyak dan paling sedikit

```python
season_calc = day_hari.groupby("season").agg({"cnt":"sum"})
season_calc.plot(kind="bar", figsize=(10, 6))
plt.title("total penyewa tiap season")
plt.xlabel("season")
plt.ylabel("jumlah penyewa")
plt.xticks(rotation=0)
plt.show()
```

- dapat disimpulkan bahwa season dengan penyewa paling banyak adalah season ke-3 (fall)
- dan penyewa paling sedikit adalah season ke-1 (springer)

---
---

> *le6 diatas grafik batang dibawah adalah notasi ilmiah pada ilmu komputer untuk angka besar yaitu = 1.000.000 (satu juta).*

- dibawah adalah tampilan penyewa paling banyak (season 3) dan penyewa paling sedikit (season 1)

```python
season_calc = day_hari.groupby("season").agg({"cnt":"sum"})
season_calc = season_calc.iloc[[0, 2]]
season_calc.plot(kind="bar", figsize=(10, 6))
plt.title("penyewa paling banyak & paling sedikit")
plt.xlabel("season")
plt.ylabel("jumlah penyewa")
plt.xticks(rotation=0)
plt.show()
```
